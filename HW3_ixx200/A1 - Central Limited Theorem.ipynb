# This Assignment Explore the Central Limit Theorem

### 1. Generate 100 Samples with Various Sample Size from 5 Different Distributions

```python
import numpy as np
import matplotlib.pylab as plt
seed = 123
%pylab inline
```

```python
normal_samples = []
poisson_samples = []
binomial_samples = []
chi_samples = []
gamma_samples = []

# Create 100 Samples for 5 Different Distribution

for i in range(100):
    
    # Each Sample has a Different Sample Size
    
    sample_size = np.random.randint(10, 2000)
    
    normal_samples.append(np.random.normal(5, 2.5, sample_size))
    poisson_samples.append(np.random.poisson(1, sample_size))
    binomial_samples.append(np.random.binomial(3, 0.5, sample_size))
    chi_samples.append(np.random.chisquare(2, sample_size))
    gamma_samples.append(np.random.gamma(9, 2, sample_size))
```

### Quick Validation

```python
print("Number of Samples of Taken From the Binomial Distribution: "+str(len(binomial_samples)))
```

```python
print("The Sample Size of a sample: " + str(len(binomial_samples[2])) + " and the mean: " + str(binomial_samples[2].mean()))
```

```python
print("The Sample Size of another sample: " + str(len(binomial_samples[20])) + " and the mean: " + str(binomial_samples[20].mean()))
```

### 2. Create Arrays of Sample Size and Corresponding Mean for Samples Taken from Each Distribution

```python
normal_array = []
poisson_array = []
binomial_array = []
chi_array = []
gamma_array = []
```

```python
# Get Each Sample Sizie and Sample Mean from Normal Distibution

def getSampleSizeMean(sample):
    
    '''
    This function returns arrays of all sample size and the corresponding sample mean
    arg: sample(list) - this is a list of all samples with various sizes taken from a distribution
    return: samplesize(list), sample mean (list)
    '''
    sample_size = []
    sample_mean = []
    
    for i in range(len(sample)):
        sample_size.append(len(sample[i]))
        sample_mean.append(sample[i].mean()) 
    
    return sample_size, sample_mean

normal_array = [getSampleSizeMean(normal_samples)[0], getSampleSizeMean(normal_samples)[1]]
poisson_array = [getSampleSizeMean(poisson_samples)[0], getSampleSizeMean(poisson_samples)[1]]
binomial_array = [getSampleSizeMean(binomial_samples)[0], getSampleSizeMean(binomial_samples)[1]]
chi_array = [getSampleSizeMean(chi_samples)[0], getSampleSizeMean(chi_samples)[1]]
gamma_array = [getSampleSizeMean(gamma_samples)[0], getSampleSizeMean(gamma_samples)[1]]
```

### 3. Plot Sample Size and Corresponding Mean

```python
plt.plot(normal_array[0], normal_array[1], "o");
plt.title("100 Sample Size vs. Sample Mean from the Normal Distribution")
plt.xlabel("Sample Size");
plt.ylabel("Sample Mean");
```

```python
plt.plot(poisson_array[0], poisson_array[1], "o");
plt.title("100 Sample Size vs. Sample Mean from the Poisson Distribution")
plt.xlabel("Sample Size");
plt.ylabel("Sample Mean");
```

```python
plt.plot(binomial_array[0], binomial_array[1], "o");
plt.title("100 Sample Size vs. Sample Mean from the Binomial Distribution")
plt.xlabel("Sample Size");
plt.ylabel("Sample Mean");
```

```python
plt.plot(chi_array[0], chi_array[1], "o");
plt.title("100 Sample Size vs. Sample Mean from the Chi Square Distribution")
plt.xlabel("Sample Size");
plt.ylabel("Sample Mean");
```

```python
plt.plot(gamma_array[0], gamma_array[1], "o");
plt.title("100 Sample Size vs. Sample Mean from the Gamma Distribution")
plt.xlabel("Sample Size");
plt.ylabel("Sample Mean");
```

### Key Observation:

- The sample mean converges to the expected value (mean of the distribution; e.g. 5 in my Normal distribution) as the sample size increases. This observation aligns with the Law of Large Number (LLN). 

- The LLN states that the average of the results obtained from a large number of trials should be close to the expected value, and will tend to become closer as more tials are performed. 

- For example, the casino may lose money in a single card game, but its earning will drive towards an expected value over a large amount of games. The casino should expect a positive profit if the games are carefully design in the favour of the firm.
